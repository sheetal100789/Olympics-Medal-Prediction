```python
#Does US shows dominance in all the sports? Or there are some sports where other countries have excelled?
olympics = pd.get_dummies(olympics,columns=['Medal'])
olympics.rename(columns={'Medal_0':'Bronze', 'Medal_1':'Gold','Medal_2':'Silver'}, inplace=True)
olympics['TotalMedals'] = olympics['Bronze'] + olympics['Gold'] + olympics['Silver']

SportByTeam = olympics.groupby(by = ['Sport','NOC'], as_index=False).sum()
```

```python
us = SportByTeam['Sport'].unique()
```

```python
topSport = pd.DataFrame()
for i in us:
    topSport = pd.concat([topSport,SportByTeam[SportByTeam['Sport'] == i].sort_values(by=['TotalMedals'],ascending= False).head(1)])


    
```

```python
topCountries = topSport[['Sport','NOC','TotalMedals']]
topCountriesnotUSA = topCountries[topCountries['NOC'] != 'USA'].sort_values(by= 'TotalMedals', ascending = False).head(15)
```
